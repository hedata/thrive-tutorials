## What are Jupyter Notebooks?
> The Jupyter Notebook allows you to create documents that contain live code, equations, visualizations and explanatory text. 

> Uses include: data cleaning and transformation, numerical simulation, statistical modeling, machine learning and much more. [Source: Jupyter](http://jupyter.org/)

The document you are reading right now is such a notebook. A notebook consists of cells. This text is written in a <b>markdown cell</b> and the <b> code cell</b> right under this cell does a little bit of python magic to show the output of a Machine Learning Method called Random Forest Classifier that separates points into 4 different categories.

````python
sample_amount=300
import warnings
warnings.filterwarnings('ignore');
%matplotlib inline
import numpy as np
from sklearn.datasets import make_blobs
from fig_code import visualize_tree
from sklearn.ensemble import RandomForestClassifier
X, y = make_blobs(n_samples=sample_amount, centers=4,
                      random_state=0, cluster_std=2.0)    
clf = RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1)
visualize_tree(clf, X, y, boundaries=False);
````

Currently this graphic shows 300 points that belong to a category and the colored areas is how the classifier would assign new points to their respective categories.
